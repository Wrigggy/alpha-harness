```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from src.data_collection.data_cleaner import load_panel
from src.backtest.long_short_backtest import long_short_backtest, plot_backtest

%matplotlib inline
```

```python
# Load data
panel = load_panel('../data/processed')
ret = panel['ret_1h']
close = panel['close']
quote_vol = panel.get('quote_volume')
```

```python
# Example: momentum factor backtest
momentum = close.pct_change(20)

result = long_short_backtest(
    factor_values=momentum,
    returns=ret,
    n_long=30,
    n_short=30,
    liquidity_filter=quote_vol,
    transaction_cost_bps=5.0,
)

print('Metrics:')
for k, v in result['metrics'].items():
    if isinstance(v, float):
        print(f'  {k}: {v:.4f}')
    else:
        print(f'  {k}: {v}')
```

```python
# Plot equity curve
plot_backtest(result['equity_curve'], result['metrics'], title='Momentum(20) Long-Short')
```
