# 01 — Data Exploration

Explore the collected Binance OHLCV data:
- Universe overview
- Data quality checks
- Missing data analysis
- Basic price/volume statistics

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from src.data_collection.data_cleaner import load_panel

%matplotlib inline
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
# Load panel data
panel = load_panel('../data/processed')
print('Fields:', list(panel.keys()))
print('Shape:', panel['close'].shape)
```

```python
# Missing data heatmap
close = panel['close']
missing_pct = close.isna().mean().sort_values(ascending=False)
print(f'Symbols with >5% missing: {(missing_pct > 0.05).sum()}')
missing_pct.head(20).plot.bar(figsize=(12, 4), title='Top 20 symbols by missing data %')
```

```python
# Return distribution
ret = panel['ret_1h']
fig, ax = plt.subplots(figsize=(10, 5))
ret.stack().hist(bins=200, ax=ax, density=True)
ax.set_xlim(-0.1, 0.1)
ax.set_title('1H Return Distribution (all symbols)')
ax.set_xlabel('Return')
```
